```python
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

iris = load_iris()
df_iris = pd.DataFrame(data= iris['data'],
                     columns= iris['feature_names'])
df_iris.head()
```

```python
df_iris.boxplot() # Muestra los datos atípicos (1.5 iqr)
```

```python
df_iris.boxplot(whis=3) # Muestra los datos extremos (3 iqr). No hay  ;)
```

```python
plt.boxplot(df_iris) # whis=3  ->  para datos extremos
plt.grid()
```

```python
# En dónde y cuáles son los datos atípicos
q1 = df_iris.quantile(0.25)
q3 = df_iris.quantile(0.75)
iqr = q3 - q1
df_iris[((df_iris < q1 - 1.5*iqr) | (df_iris > q3 + 1.5*iqr)).values]
```
