```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def anti_yj(x_prime, corderita):
    x_prime = np.array(x_prime)
    x = np.zeros_like(x_prime)

    # x' >= 0
    pos = x_prime >= 0
    if corderita != 0:
        x[pos] = np.power(corderita * x_prime[pos] + 1, 1 / corderita) - 1
    else:
        x[pos] = np.exp(x_prime[pos]) - 1

    # x' < 0
    neg = x_prime < 0
    if corderita != 2:
        x[neg] = 1 - np.power(-(2 - corderita) * x_prime[neg] + 1, 1 / (2 - corderita))
    else:
        x[neg] = -np.exp(-x_prime[neg]) + 1

    return x
```

```python
x_prime_vals = np.linspace(0.001, 0.999, 500)
lambdas = [-3, -2, -1, 0, 0.5, 1, 2, 3]

plt.figure(figsize=(10, 6))

for lmbd in lambdas:
    x_vals = anti_yj(x_prime_vals, lmbd)
    plt.plot(x_prime_vals, x_vals, label=f"lambda = {lmbd}")

plt.title("Curvas inversas de Yeo-Johnson")
plt.xlabel("x'")
plt.ylabel("x")
plt.legend()
plt.grid(True)
plt.show()
```
