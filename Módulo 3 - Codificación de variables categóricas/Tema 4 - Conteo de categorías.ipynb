````python
import pandas as pd
import numpy as np
ruta='/content/drive/MyDrive/Colab Notebooks/'

data=pd.read_csv(ruta+'temps.txt',sep='\s+',header=None) # Fuente: http://astro.iam.udg.mx/Estacion/NOAAPRMO.TXT
                                                         #         http://astro.iam.udg.mx/Estacion/NOAAMO.TXT
data.columns=['Temperatura máxima','Temperatura mínima','HDD','CDD','Rain']
d=['Sábado','Domingo','Lunes','Martes','Miércoles','Jueves','Viernes']
dias=[d[i%7] for i in range(len(data))]
catego=['Soleado','Mayormente soleado','Parcialmente nublado','Mayormente nublado','Nublado','Lluvioso']
prono=[catego[-1] if data['Rain'][i]>0 else catego[int(round(1.2*(data['HDD'][i]-data['CDD'][i])+2))] for i in range(len(data))]
df=pd.DataFrame(columns=['Días','Pronóstico','Temperatura máxima','Temperatura mínima'],index=range(len(data)))
df.iloc[:,2:4]=data.iloc[:,0:2]
df['Días']=dias
df['Pronóstico']=prono

txt1="Pedro Perez Pereira, pobre pintor portuguez, pinta preciosos paisajes para poder partir para Paris."
txt2="Preciosos gatos y perros con mancha quiero. Partir de izquierda a derecha para poder llegar a algun lugar."
txt3="En un lugar de la Mancha, cuyo nombre no quiero acordarme..."
````

````python
# Expresiones regulares
import re
````

[Expresiones regulares re](https://docs.python.org/3/library/re.html)

[RegEx](https://pypi.org/project/regex/)
```python
'\W'  # caracteres especiales
'\w'  # caracter alfanumérico
'\d'  # caracter numérico
'\D'  # caracter diferente al numérico
'\s'  # espacio
'\W+' # al menos un caracter especial
'a*'  # 0 o más 'a'
'a?'  # 0 o 1 'a'
'perr[ao]' # disyunción
```

````python
listaPalabras=re.split('\W+',(txt1+txt2+txt3).lower())
listaPalabras
````

````python
palabras=set(listaPalabras)
palabras.remove('')
palabras=list(palabras)
palabras
````

````python
# Contamos las palabras para cada variable
conteo=[[],[],[]]
for palabra in palabras:
  conteo[0].append(re.split('\W+',txt1.lower()).count(palabra))
  conteo[1].append(re.split('\W+',txt2.lower()).count(palabra))
  conteo[2].append(re.split('\W+',txt3.lower()).count(palabra))
pd.DataFrame(conteo,columns=palabras,index=['txt1','txt2','txt3'])
````

````python
#@title Usando librerías
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
````

````python
txt1,txt2,txt3
````

````python
codificador=CountVectorizer() # stop_words=['el','la','para','a',...]
m_conteo=codificador.fit_transform([txt1,txt2,txt3])
m_conteo
````

````python
codificador.get_feature_names_out()
````

````python
set(palabras)-set(codificador.get_feature_names_out())
````

````python
pd.DataFrame(m_conteo.toarray(),columns=codificador.get_feature_names_out(),index=['txt1','txt2','txt3'])
````

````python
codificacion=CountVectorizer(ngram_range=(1,3))
m_conteo=codificacion.fit_transform([txt1,txt2,txt3])
df_conteo=pd.DataFrame(m_conteo.toarray(),columns=codificacion.get_feature_names_out(),index=['txt1','txt2','txt3'])
df_conteo
````
